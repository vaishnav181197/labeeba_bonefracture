```python
import os
import cv2
import tensorflow as tf
from tensorflow import keras
```

```python
model = tf.keras.models.load_model('facedetcetor.h5')

# Show the model architecture
model.summary()

```

```python
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt

h, w = 100, 100

img = image.load_img('10.jpg', target_size=(100,100,3))
imag = cv2.imread('10.jpg')
gray=cv2.cvtColor(np.float32(img),cv2.COLOR_BGR2GRAY)
out = np.expand_dims(gray,axis=0)
final_img = out/255.0

pred = model.predict(final_img)
max_index=np.argmax(pred[0])
print(max_index)
emotions= ('fractured','not fractured')
predicted_emotion=emotions[max_index]
print(pred[0])
print(predicted_emotion)

plt.imshow(cv2.cvtColor(imag, cv2.COLOR_BGR2RGB))
# as opencv loads in BGR format by default, we want to show it in RGB.
plt.show()
```
